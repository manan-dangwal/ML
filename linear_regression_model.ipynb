```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['auto-mpg.csv']))

df = df.applymap(lambda x: np.nan if x == '?' else x).dropna()

print(df)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

print(train_set)
```

```python
import pandas as pd
import numpy as np
import math
import operator
import matplotlib.pyplot as plt
%matplotlib inline



# collecting vectors
W = train_set['weight'].values
HP = train_set['horsepower'].values
HP = [int(j) for j in HP]   # value to be predicted
CY = train_set['cylinders'].values
D = train_set['displacement'].values
MPG = train_set['mpg'].values
A = train_set['acceleration'].values

```

```python
# finding parameters using normal equation algorithm

X = np.array([W, CY, D, MPG, A]).transpose()

y = np.array([HP]).transpose()

p = np.dot(X.transpose(), X)
p = np.linalg.inv(p)   # inverting the matrix
q = np.dot(p, X.transpose())
theta = np.dot(q, y)    # parameters
```

```python
# testing

Wt = test_set['weight'].values
HPt = test_set['horsepower'].values
HPt = [int(j) for j in HPt]   # value to be predicted
CYt = test_set['cylinders'].values
Dt = test_set['displacement'].values
MPGt = test_set['mpg'].values
At = test_set['acceleration'].values

Xt = np.array([Wt, CYt, Dt, MPGt, At]).transpose()
yt = np.array([HPt]).transpose()


predicted = np.dot(Xt, theta)

```

```python
c = np.linspace(0, 250, 5)
d = c

plt.plot(c, d)
plt.scatter(yt, predicted, c='#ef5423')

plt.xlabel('Actual Horsepower')
plt.ylabel('Predicted Horsepower')
plt.legend()
plt.show()
```
