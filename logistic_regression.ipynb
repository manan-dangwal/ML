```python
import matplotlib.pyplot as plt

# Sample data (replace this with your own dataset)
x1 = df["X_1"].values
x2 = df["X_2"].values
y = df["y"].values


# Create a scatter plot
plt.figure(figsize=(8, 6))

# Separate data points for class 0 and class 1
class_0_x1 = [x1[i] for i in range(len(x1)) if y[i] == 0]
class_0_x2 = [x2[i] for i in range(len(x2)) if y[i] == 0]
class_1_x1 = [x1[i] for i in range(len(x1)) if y[i] == 1]
class_1_x2 = [x2[i] for i in range(len(x2)) if y[i] == 1]

# Plot class 0 with circles and class 1 with crosses
plt.scatter(class_0_x1, class_0_x2, c='blue', marker='o', label='Class 0')
plt.scatter(class_1_x1, class_1_x2, c='red', marker='x', label='Class 1')

# v = np.linspace (-1, 1, 1000)
# theta = np.array([1, 1, 1])
# w = (theta[0] + theta[1] * v) * (-1)
# plt.scatter(v, w, s=0.25, marker='o', color='green')

# Customize the plot
plt.xlabel('Parameter 1')
plt.ylabel('Parameter 2')
plt.legend()
plt.title('2D Scatter Plot with Circles (0) and Crosses (1)')

# Show the plot
plt.show()
```

```python
plt.scatter(class_0_x1, class_0_x2, c='blue', marker='o', label='Class 0')
plt.scatter(class_1_x1, class_1_x2, c='red', marker='x', label='Class 1')

import numpy as np

a = 0.5 # learning rate
def sigmoid(z):
    # return 1 / (1 + np.exp(-z))

    if z >= 0:
      return 1
    else:
      return 0



v = np.linspace (-1, 1, 1000)
theta_0 = 1
theta_1 = 2
theta_2 = 3

# theta = np.array([1, 1, 1])

# w = -1 * (theta_0 + theta_1 * v) / theta_2 #plots the line
# plt.scatter(v, w, s=0.25, marker='o', color='black')

k = 0
while (k < 20):
  err = []
  for i in range(len(x1)):
    theta_x_i = theta_0 + theta_1 * x1[i] + theta_2 * x2[i]
    h_theta_x_i = sigmoid(theta_x_i)
    err.append(y[i] - h_theta_x_i)

  err_sum = 0
  for j in range(len(err)):
    err_sum += err[j]

  err = np.array(err)

  theta_0 = theta_0 + a * err_sum
  theta_1 = theta_1 + a * np.dot(err.T, x1)
  theta_2 = theta_2 + a * np.dot(err.T, x2)

  k = k+1


w = -1 * (theta_0 + theta_1 * v) / theta_2 #plots the line
plt.scatter(v, w, s=0.15, marker='o', color='teal')



plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
